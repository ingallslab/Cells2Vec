```python
import cv2
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import numpy as np
import numpy as np
import matplotlib.pyplot as plt
```

```python

input_data=np.load('batch.npy')
prediction_data=np.load('pred.npy')
print(input_data.shape), print(prediction_data.shape)
```

```python

fig, axes = plt.subplots(nrows=19, ncols=2, figsize=(24, 72))

for i in range(19):

    axes[i, 0].imshow(input_data[0, :, :, i], cmap='hot')
    axes[i, 0].set_title(f'Input Heatmap (Channel {i+1})')
    axes[i, 0].axis('off')
    axes[i, 1].imshow(prediction_data[0, :, :, i], cmap='hot')
    axes[i, 1].set_title(f'Prediction Heatmap (Channel {i+1})')
    axes[i, 1].axis('off')
plt.subplots_adjust(hspace=0.5)
plt.show()
```


 Use below code to generate images(Not heatmaps) using CV2

```python
for i in range(19):
    cv2.imshow(f'Orig Channel: {i + 1}', cv2.resize(input_data[0][:, :,i], None, fx = 5, fy = 5))
    cv2.imshow(f'Pred Channel: {i + 1}', cv2.resize(prediction_data[0][:, :,i], None, fx = 5, fy = 5))
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
